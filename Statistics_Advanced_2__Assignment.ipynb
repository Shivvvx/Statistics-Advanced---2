**Q1.What is hypothesis testing in statistics?**
- Ans.1 Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data. It starts with two assumptions: the **null hypothesis (H₀)**, which states that there is no effect or no difference, and the **alternative hypothesis (H₁ or Ha)**, which suggests that there is a significant effect or difference. By analyzing the sample data, statisticians use test statistics (such as *t*, *z*, or chi-square values) to determine whether the evidence is strong enough to reject the null hypothesis. The decision is made by comparing the *p-value* (probability of observing the result if the null is true) with a chosen significance level (α, often 0.05). If the p-value is less than α, the null hypothesis is rejected in favor of the alternative. In simple terms, hypothesis testing provides a structured way to evaluate claims or theories using data rather than assumptions.


**Q2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?**
- Ans. The **null hypothesis (H₀)** is a statistical assumption that there is **no effect, no difference, or no relationship** between variables in the population. It represents the default or status quo claim that researchers test against. For example, in testing a new medicine, the null hypothesis may state that the medicine has **no effect** compared to the existing treatment.

On the other hand, the **alternative hypothesis (H₁ or Ha)** is the statement that contradicts the null. It suggests that there **is an effect, a difference, or a relationship** in the population. Using the same example, the alternative hypothesis would state that the new medicine **does have an effect** compared to the existing treatment.

The key difference is that the null hypothesis assumes no change or effect, while the alternative hypothesis assumes the presence of change or effect. Hypothesis testing works by using data to decide whether to reject the null in favor of the alternative, or to fail to reject the null if there isn’t enough evidence.


**Q3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?**
- Ans.   The **significance level (α)** in hypothesis testing is the threshold used to decide whether to reject the null hypothesis. It represents the maximum probability of making a **Type I error**, which means rejecting the null hypothesis when it is actually true. Common values of α are **0.05 (5%)**, **0.01 (1%)**, or **0.10 (10%)**, depending on how strict the test needs to be.

The role of the significance level is crucial: after calculating the **p-value** from the sample data, we compare it with α.

* If the **p-value ≤ α**, we reject the null hypothesis, concluding that the evidence is strong enough to support the alternative hypothesis.
* If the **p-value > α**, we fail to reject the null hypothesis, meaning the data does not provide sufficient evidence against it.

In short, the significance level acts as a **cutoff point** that helps control the risk of wrongly rejecting the null hypothesis and ensures that decisions in hypothesis testing are made with a predefined level of confidence.


**Q4.What are Type I and Type II errors? Give examples of each ?**
-  Ans.  In hypothesis testing, **Type I and Type II errors** are the two possible mistakes that can occur when making decisions about hypotheses:

* **Type I Error (False Positive):** This happens when we **reject the null hypothesis (H₀) even though it is actually true**. In other words, we conclude there is an effect or difference when in reality there isn’t.
  **Example:** A medical test concludes that a patient has a disease when the patient is actually healthy. Here, the null hypothesis (patient is healthy) is wrongly rejected.

* **Type II Error (False Negative):** This occurs when we **fail to reject the null hypothesis even though it is false**. In this case, we miss detecting a real effect or difference.
  **Example:** A medical test concludes that a patient does not have a disease when the patient actually does. Here, the null hypothesis (patient is healthy) is wrongly accepted because the test missed the real condition.


* **Type I Error = “False alarm.”**
* **Type II Error = “Missed detection.”**



**Q5. What is the difference between a Z-test and a T-test? Explain when to use each?**
- Ans. A **Z-test** and a **T-test** are both statistical tests used to compare sample data with population data or between groups, but they differ mainly in their assumptions and sample size requirements.

* **Z-test:**
  A Z-test is used when the population **standard deviation (σ) is known** and the sample size is **large (n > 30)**. It assumes the data follows a **normal distribution**. It is commonly used for testing population means, proportions, and differences between means when conditions are ideal.
  **Example:** Checking if the average height of 1000 students differs from a known population mean when the population standard deviation is available.

* **T-test:**
  A T-test is used when the population **standard deviation is unknown** and the sample size is **small (n < 30)**. It uses the **sample standard deviation (s)** as an estimate of σ and relies on the **t-distribution**, which accounts for extra uncertainty with small samples.
  **Example:** Testing if the average exam score of a class of 20 students differs from the expected population mean when σ is not known.

* Use a **Z-test** for **large samples** with **known σ**.
* Use a **T-test** for **small samples** with **unknown σ**.

```python
# Qest 7. 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# solution:

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Step 1: Define hypotheses
# H0: μ = 50 (population mean is 50)
# H1: μ ≠ 50 (population mean is not 50)

mu_0 = 50   # hypothesized population mean
alpha = 0.05  # significance level

# Step 2: Compute sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Step 3: Compute Z statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Step 4: Compute p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Decision
if p_value < alpha:
    decision = "Reject the null hypothesis (H0)"
else:
    decision = "Fail to reject the null hypothesis (H0)"

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Sample Std Dev = {sample_std:.4f}")
print(f"Z-Statistic = {z_stat:.4f}")
print(f"P-Value = {p_value:.4f}")
print(f"Decision: {decision}")

```

```python
# Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# Solution:
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # population mean and std deviation
n = 100             # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)  # 1.96 for 95%
margin_of_error = z_critical * se
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"95% Confidence Interval = ({ci_lower:.4f}, {ci_upper:.4f})")

# Step 5: Plot histogram of the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='--', label=f"Upper 95% CI = {ci_upper:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

# Solution:
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Step 1: Calculate mean and std deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std deviation

    # Step 2: Compute Z-scores
    z_scores = (data - mean) / std

    # Step 3: Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label="Mean (0)")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(42)
sample_data = np.random.normal(100, 15, 200)  # normal dist with mean=100, std=15
z_scores = calculate_z_scores(sample_data)

print("First 10 Z-scores:", z_scores[:10])
```
